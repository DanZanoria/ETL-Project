# Transform

#### World happiness 2015

```python
world_happiness_index_2015_df.columns
```

```python
del world_happiness_index_2015_df['Region']
del world_happiness_index_2015_df['Standard Error']
world_happiness_index_2015_df.rename(columns = {'Country':'country',
                                                    'Happiness Rank':'happiness_rank',
                                                    'Happiness Score':'happiness_score',
                                                    'Economy (GDP per Capita)':'gdp_per_capita',
                                                    'Family':'family',
                                                    'Health (Life Expectancy)':'life_expectancy',
                                                    'Freedom':'freedom',
                                                    'Generosity':'generosity',
                                                    'Trust (Government Corruption)':'trust_government',
                                                    'Dystopia Residual':'dystopia_residual'}, inplace = True)
world_happiness_index_2015_df['year']=2015
world_happiness_index_2015_df.head()
```

#### World happiness 2016

```python
world_happiness_index_2016_df.columns
```

```python
del world_happiness_index_2016_df['Region']
del world_happiness_index_2016_df['Lower Confidence Interval']
del world_happiness_index_2016_df['Upper Confidence Interval']
world_happiness_index_2016_df.rename(columns = {'Country':'country',
                                                    'Happiness Rank':'happiness_rank',
                                                    'Happiness Score':'happiness_score',
                                                    'Economy (GDP per Capita)':'gdp_per_capita',
                                                    'Family':'family',
                                                    'Health (Life Expectancy)':'life_expectancy',
                                                    'Freedom':'freedom',
                                                    'Generosity':'generosity',
                                                    'Trust (Government Corruption)':'trust_government',
                                                    'Dystopia Residual':'dystopia_residual'}, inplace = True)
world_happiness_index_2016_df['year']=2016
world_happiness_index_2016_df.head()
```

#### World happiness 2017

```python
world_happiness_index_2017_df.columns
```

```python
del world_happiness_index_2017_df['Whisker.high']
del world_happiness_index_2017_df['Whisker.low']
world_happiness_index_2017_df.rename(columns = {'Country':'country',
                                                    'Happiness.Rank':'happiness_rank',
                                                    'Happiness.Score':'happiness_score',
                                                    'Economy..GDP.per.Capita.':'gdp_per_capita',
                                                    'Family':'family',
                                                    'Health..Life.Expectancy.':'life_expectancy',
                                                    'Freedom':'freedom',
                                                    'Generosity':'generosity',
                                                    'Trust..Government.Corruption.':'trust_government',
                                                    'Dystopia.Residual':'dystopia_residual'}, inplace = True)
world_happiness_index_2017_df = world_happiness_index_2017_df[['country', 'happiness_rank', 'happiness_score', 'gdp_per_capita',
       'family', 'life_expectancy', 'freedom', 'trust_government', 'generosity', 'dystopia_residual']]
world_happiness_index_2017_df['year']=2017
world_happiness_index_2017_df.head()
```

```python
# appending all the world happiness datasets
world_happiness_index_df = world_happiness_index_2015_df.append(world_happiness_index_2016_df, ignore_index = True)
world_happiness_index_df = world_happiness_index_df.append(world_happiness_index_2017_df, ignore_index = True)
world_happiness_index_df.head(-5)
```

```python
# Creating a dataframe for standardized county names for each country in world_happiness_index_df
entity1= list(set(world_happiness_index_df["country"]))
standard_names1 = coco.convert(names=entity1, to='name_short')
standardized_names1_df = pd.DataFrame({"country" : entity1,
                                      "standard_names": standard_names1}).sort_values(by = "country").reset_index(drop = True)
standardized_names1_df.head()
```

```python
# joining with world_happiness_index_df to get the standardized names
world_happiness_index_df = pd.merge(world_happiness_index_df, standardized_names1_df, how = "left", left_on='country', right_on='country')
del world_happiness_index_df['country']
world_happiness_index_df.rename(columns = {'standard_names': 'country'}, inplace =True)
world_happiness_index_df= world_happiness_index_df[['country', 'happiness_rank', 'happiness_score', 'gdp_per_capita',
       'family', 'life_expectancy', 'freedom', 'trust_government',
       'generosity', 'dystopia_residual', 'year']]
world_happiness_index_df.head(-5)
```

#### World alcohol consumption per capita

```python
alcoholSubstanceAbuse_df.columns
```

```python
del alcoholSubstanceAbuse_df['Indicator']
alcoholSubstanceAbuse_df.rename(columns = {'Location' : 'country', 'Period':'year',
                                                'Dim1':'gender','First Tooltip': 'per_capita_consumption'}, inplace = True)
alcoholSubstanceAbuse_df.head()
```

```python
# Creating a dataframe for standardized county names for each country in alcoholSubstanceAbuse_df
entity2= list(set(alcoholSubstanceAbuse_df["country"]))
standard_names2 = coco.convert(names=entity2, to='name_short')
standardized_names2_df = pd.DataFrame({"country" : entity2,
                                      "standard_names": standard_names2}).sort_values(by = "country").reset_index(drop = True)
standardized_names2_df.head()
```

```python
# joining with alcoholSubstanceAbuse_df to get the standardized names
alcoholSubstanceAbuse_df = pd.merge(alcoholSubstanceAbuse_df, standardized_names2_df, how = "left", left_on='country', right_on='country')
del alcoholSubstanceAbuse_df['country']
alcoholSubstanceAbuse_df.rename(columns = {'standard_names': 'country'}, inplace =True)
alcoholSubstanceAbuse_df= alcoholSubstanceAbuse_df[['country', 'year', 'gender', 'per_capita_consumption']]
alcoholSubstanceAbuse_df.head(-5)
```

#### World antidepressant consumption per 1000 people per day

```python
antidepressant_consumption_df.columns
```

```python
antidepressant_consumption_df['notes']= antidepressant_consumption_df['Data include drugs dispensed in hospitals']+ \
                                        antidepressant_consumption_df['Data include non-reimbursed drugs']+ \
                                        antidepressant_consumption_df['Data include OTC drugs']+\
                                        antidepressant_consumption_df['Notes']
del antidepressant_consumption_df['Data include drugs dispensed in hospitals']
del antidepressant_consumption_df['Data include non-reimbursed drugs']
del antidepressant_consumption_df['Data include OTC drugs']
del antidepressant_consumption_df['Notes']
antidepressant_consumption_df.rename(columns = {'Country': 'country',
                                                     'Defined daily dosage per 1 000 inhabitants per day': 'daily_dosage_per_1k',
                                                     'Year': 'year'}, inplace=True)
antidepressant_consumption_df.head()
```

```python
# Creating a dataframe for standardized county names for each country in antidepressant_consumption_df
entity3= list(set(antidepressant_consumption_df["country"]))
standard_names3 = coco.convert(names=entity3, to='name_short')
standardized_names3_df = pd.DataFrame({"country" : entity3,
                                      "standard_names": standard_names3}).sort_values(by = "country").reset_index(drop = True)
standardized_names3_df.head()
```

```python
# joining with antidepressant_consumption_df to get the standardized names
antidepressant_consumption_df = pd.merge(antidepressant_consumption_df, standardized_names3_df, how = "left", left_on='country', right_on='country')
del antidepressant_consumption_df['country']
antidepressant_consumption_df.rename(columns = {'standard_names': 'country'}, inplace =True)
antidepressant_consumption_df= antidepressant_consumption_df[['country', 'daily_dosage_per_1k', 'year', 'notes']]
antidepressant_consumption_df.head(-5)
```
